### TPC1: Jogo 21 fósforos

```python
def primeiro():
    a=input ("Quem inicia o jogo?")
    if a=="computador":
        computador()
    elif a=="jogador":
        jogador()
    else:
        print("Escolha jogador ou computador")
        primeiro()

#jogador a iniciar o jogo
def jogador():
    S=21
    while (S>1):
        d=input ("Escolha um número entre 1 e 4")
        d=int(d)
        #jogador a escolher um numero (falta adicionar "erro" caso d não seja um número - jogador pode escolher uma letra?)
        if d>=1 and d<=4:
            print("Jogador retirou",d,"fósforos")
        else:
            print("Tente novamente.")
            continue

        S=S-d
        #numero de fosforos após a jogada
        e=5-d 
        #computador a completar 5
        print("Computador retirou",e)
        S=S-e
        print("Sobram",S,"fósforos!")

    if (S==1):
        print("O jogador perdeu o jogo.")

            
#computador a iniciar o jogo
def computador():
    S=21

    while S==21:
        import random
        b= random.randint(1,4) # primeira jogada do computador é aleatoria
        print("Computador retirou",b)
        S=S-b
        print("Sobram",S,"fósforos!")

        
        c= input("Escolha um número entre 1 e 4")
        c= int(c) #jogador joga em 2º lugar

        if c>=1 and c<=4:
            print("Jogador retirou",c,"fósforos")
            S=S-c
        
        else:
            print("Tente novamente.")
            continue

#3 casos possiveis após a 2ª jogada do jogador
# CASO 1
        while (b+c)==5:

            import random
            b= random.randint(1,4) #computador joga aleatoriamente de novo
            print("Computador retirou",b)
            S=S-b
            print("Sobram",S,"fósforos!")
                
            c= input("Escolha um número entre 1 e 4")
            c= int(c)

            if c>=1 and c<=4:
                print("Jogador retirou",c,"fósforos")
                S=S-c
        
            else:
                print("Tente novamente.")
                continue
                
        if S==1:
            print("Sobrou apenas 1 fósforo!")
            print("O jogador ganhou o jogo :D")


        #outros casos na última jogada
        if S==0:
            print("Tirou o último fósforo. O computador ganhou o jogo.")
            #se o jogador tirar o último fósforo junto com outros

        if S-c<0:
            print("Não pode tirar fósforos inexistentes.")
            break
            #se na última jogada o jogador tirar mais fósforos do que os possíveis


# CASO 2
        if (b+c)<5:
            b=5-(b+c)
            print("Computador retirou",b)
            S=S-b
            print("Sobram",S,"fósforos!")

            while S>1:
                c= input("Escolha um número entre 1 e 4")
                c= int(c)

                if c>=1 and c<=4:
                    print("Jogador retirou",c,"fósforos")
                    S=S-c
        
                else:
                    print("Tente novamente.")
                    continue
        
                b=5-c #depois do erro do jogador, o computador completa sempre 5
                print("Computador retirou",b)
                S=S-b
                print("Sobram",S,"fósforos!")

            if (S<=1):
                print("O jogador perdeu o jogo.") 

# CASO 3
#se, por exemplo, o computador começar por tirar 4 e o jogador também tire 4, o computador tira 2 para completar 10 e fechar duas rodadas (5+5)

        else: #b+c>5
            b=10-(b+c)
            print("Computador retirou",b)
            S=S-b
            print("Sobram",S,"fósforos!")

            while S>1:

                c= input("Escolha um número entre 1 e 4")
                c= int(c)

                if c>=1 and c<=4:
                    print("Jogador retirou",c,"fósforos")
                    S=S-c
        
                else:
                    print("Tente novamente.")
                    continue
                
                b=5-c #depois do erro do jogador, o computador completa sempre 5
                print("Computador retirou",b)
                S=S-b
                print("Sobram",S,"fósforos!")

            if (S<=1):
                print("O jogador perdeu o jogo.") 




primeiro()
```
